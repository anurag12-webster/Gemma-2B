```python
from transformers import MBartForConditionalGeneration, MBart50TokenizerFast
model = MBartForConditionalGeneration.from_pretrained("facebook/mbart-large-50-one-to-many-mmt")
tokenizer = MBart50TokenizerFast.from_pretrained("facebook/mbart-large-50-one-to-many-mmt", src_lang="en_XX")
```

```python
with open("transcription.txt", "r") as file:
  input_text = file.readlines()
```

```python
model_inputs= tokenizer(input_text, return_tensors="pt", padding=True, truncation=True)
```

```python
generated_tokens = model.generate(
    **model_inputs,
    forced_bos_token_id=tokenizer.lang_code_to_id["hi_IN"]
)
```

```python
translation = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
print(translation)
```

```python
with open("hindi_transcripts.txt", "w") as file:
  for sentence in translation:
    file.write(sentence + "\n")
```

```python
from TTS.tts.configs.xtts_config import XttsConfig
from TTS.tts.models.xtts import Xtts

from IPython.display import Audio
from scipy.io.wavfile import write
```

```python
config = XttsConfig()
config.load_json("./XTTS-v2/config.json")
model = Xtts.init_from_config(config)
model.load_checkpoint(config, checkpoint_dir="./XTTS-v2/")
model.cuda()
```

```python
text_to_speak = '''दिनचर्या के बारे में कैसे बात करें? मेरा दिनचर्या कुछ भी विशेष नहीं है. मैं प्रायः सुबह 7 बजे के आसपास सुबह उठता हूँ. नाश्ता मेरे लिए महत्वपूर्ण नहीं है, इसलिए मैं प्रायः कॉफी का एक कप पीता हूँ. कभी-कभी, हालांकि, मेरे पास एक सेब होगा. नाश्ता के बाद एक शौच होता है. जब मैं कपड़े पहनता हूँ, तो मैं काम पर जाता हूँ. सुबह के दौरान काम हमेशा एक ही होता है. मैं ई-मेलों को जवाब देता हूँ और बैठकें स्थापित करता हूँ. दोपहर के समय के आसपास, मैं वास्तव में भूख महसूस करने लगता हूँ'''
reference_audios = ["audio.wav"]

outputs = model.synthesize(
    text_to_speak,
    config,
    speaker_wav=reference_audios,
    gpt_cond_len=3,
    language="en",
)
```

```python
gpt_cond_latent, speaker_embedding = model.get_conditioning_latents(audio_path=reference_audios)

outputs = model.inference(
    text=text_to_speak,
    gpt_cond_latent=gpt_cond_latent,
    speaker_embedding=speaker_embedding,
    language="en",
    enable_text_splitting=True
)
```

```python
# cloned voice output
Audio(data=outputs["wav"], rate = 24000)
```

```python
output_file_path = f'output_audio.wav'
write(output_file_path, 24000, outputs['wav'])
```
